```python
import pandas as pd
```

```python
cyberlib = pd.read_csv('cyberlib 1-15 feb.csv')
```

```python
data = cyberlib[['Content']]
```

```python
data['Content'] = data['Content'].str.lower()
data['Content'] = data['Content'].str.replace(r'(https?:\/\/)?([\da-z\.-]+)\.([a-z\.]{2,6})([\/\w \.-]*)', '')
data['Content'] = data['Content'].str.replace(r'[^\w\s\-]', '')
data['Content'] = data['Content'].str.replace(r'\d+', '')
data['Content'] = data['Content'].str.replace(r'[\-]', ' ')
data['Content'] = data['Content'].str.replace(r'\s{2,}', ' ')
```

```python
konotasi = pd.read_csv('konotasi.csv')
```

```python
konotasi
```

```python
kon = konotasi[['Konotasi negatif', 'Konotasi positif']]
```

```python
kon
```

```python
kon['Konotasi negatif'] = kon['Konotasi negatif'].str.lower()
kon['Konotasi positif'] = kon['Konotasi positif'].str.lower()
```

```python
kon
```

```python
data['Positif'] = ''
data['Negatif'] = ''
```

```python
import nltk
from nltk import sent_tokenize, word_tokenize
from nltk.tokenize import MWETokenizer
from nltk.corpus import stopwords
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

```python
for i, w in enumerate(data['Content']):
    pos = 0
    neg = 0
    
    hasil = word_tokenize(w)
    
    for j in hasil:
        if kon['Konotasi negatif'].str.contains(j).any() == True:
            neg += 1
        elif kon['Konotasi positif'].str.contains(j).any() == True:
            pos += 1
        else:
            continue
    
    data.at[i, 'Konotasi negatif'] = neg
    data.at[i, 'Konotasi positif'] = pos
```

```python
data
```

```python
data['Total'] = data['Konotasi positif'] - data['Konotasi negatif']
```

```python
data
```

```python
data.sort_values(['Total'], ascending=False)
```
