```python
import pandas as pd
```

```python
cyberlib = pd.read_csv('cyberlib 16-28 feb.csv')
```

```python
display(cyberlib.columns)
```

```python
cyberlib['Combined'] = cyberlib['Title'].astype(str) + " " + cyberlib['Content'].astype(str)
```

```python
data = cyberlib[['Combined']]
```

```python
data.columns
```

```python
data.head(3)
```

```python
data['Combined'] = data['Combined'].str.lower()
data['Combined'] = data['Combined'].str.replace(r'(https?:\/\/)?([\da-z\.-]+)\.([a-z\.]{2,6})([\/\w \.-]*)', '')
data['Combined'] = data['Combined'].str.replace(r'[^\w\s\-]', '')
data['Combined'] = data['Combined'].str.replace(r'\d+', '')
data['Combined'] = data['Combined'].str.replace(r'[\-]', ' ')
data['Combined'] = data['Combined'].str.replace(r'\s{2,}', ' ')
```

```python
data
```

```python
import nltk
from nltk import sent_tokenize, word_tokenize
from nltk.tokenize import MWETokenizer
from nltk.corpus import stopwords
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

```python
words = []

stop_words = set(stopwords.words("indonesian"))

factory = StemmerFactory()
stemmer = factory.create_stemmer()

for i, row in data.iterrows():
    result = stemmer.stem(data.at[i, 'Combined'])
    data.at[i, 'Combined'] = result
```
